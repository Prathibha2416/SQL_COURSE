```python
#import libraries 


import sqlite3
import pandas as pd
import numpy as np
import prettytable


#set table formate to DEFAULT

prettytable.DEFAULT = 'DEFAULT'

#create connection object 

conn = sqlite3.connect("students_report_card.db")

#create cursor object 

cursor_object = conn.cursor()

#load extension to SQL
%load_ext sql

#connect to database 

%sql sqlite:///students_report_card.db

#read csv file to df

df = pd.read_csv("ChicagoPublicSchools.csv")

#store the dataset in a table "CHICAGO_STUDENT_REPORT_CARD_DATA" 

df.to_sql("CHICAGO_STUDENT_REPORT_CARD_DATA", conn, if_exists = 'replace', index = False)
```

```python
#What are the top 10 schools with the highest "Average Student Attendance"?
```

```python
%%sql 
SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE
FROM CHICAGO_STUDENT_REPORT_CARD_DATA
ORDER BY AVERAGE_STUDENT_ATTENDANCE DESC
LIMIT 10;
```

```python
#Retrieve the list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance
```

```python
%%sql 
select NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE 
from CHICAGO_STUDENT_REPORT_CARD_DATA
order by AVERAGE_STUDENT_ATTENDANCE
limit 5;
```

```python
#Now remove the '%' sign from the above result set for Average Student Attendance column
```

```python
%%sql 
select NAME_OF_SCHOOL, replace(AVERAGE_STUDENT_ATTENDANCE, '%', "")
from CHICAGO_STUDENT_REPORT_CARD_DATA
order by AVERAGE_STUDENT_ATTENDANCE
limit 5;
```

```python
#Which Schools have Average Student Attendance lower than 70%?
```

```python
%%sql 
select NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE
from CHICAGO_STUDENT_REPORT_CARD_DATA
where AVERAGE_STUDENT_ATTENDANCE < 70
order by AVERAGE_STUDENT_ATTENDANCE;
```

```python
#Get the total College Enrollment for each Community Area
```

```python
%%sql
select COMMUNITY_AREA_NAME, sum(COLLEGE_ENROLLMENT) as total_college_enrollments
from CHICAGO_STUDENT_REPORT_CARD_DATA
group by COMMUNITY_AREA_NAME;
    
    
    
```
